## MOHO
The moho is the boundary between the crust and the mantle.

It's whre rocks change from the relatively light, silica-rich types to denser, magnesium and iron-rich ones.

Typical depth:

Continents= 30-45 km.
Oceans= 5-10 km.
Under mountain belts= can reach 50-60 km.

The crust above the moho is often treated as elastic or brittle-elastic while below it materials can be ductile or viscoelastic, depending on temperature.


## LAB (Lithopshre-Asthresophere Boundary)

the lab lies deepere - it marks the base of the mechanically strong lithopshre and the start of the weaker asthenosphere.

LAB defines the bottom of the region that meaningfully participates in postseismic relaxation.

Everything above it (lithosphere) can store elastic stress; everthing below (asthenopshere) is so weak it relaxes stresss almost immediately.
